### Exercise 1 - Importing matplotlib and pyplot

```python
# Import matplotlib.pyplot with the alias plt
import matplotlib.pyplot as plt

days = np.arange(1, 11)  # Days from 1 to 10
prices = np.array([100, 102, 101, 105, 107, 110, 108, 111, 115, 120])  # Example stock prices
# Plot the price of stock over time
plt.plot(days, prices, color="red", linestyle="--")

# Display the plot
plt.show()
```

### Exercise 2 - Adding axis labels and titles

```python
# Plot price as a function of time
plt.plot(days, prices, color="red", linestyle="--")

# Add x and y labels
plt.xlabel('Days')
plt.ylabel('Prices, $')

# Add plot title
plt.title('Company Stock Prices Over Time')

# Show plot
plt.show()
```

### Exercise 3 - Multiple lines on the same plot

```python
# Plot two lines of varying colors 
days = np.arange(1, 11)  # Days from 1 to 10
prices1 = np.array([100, 102, 101, 105, 107, 110, 108, 111, 115, 120])  # Example stock prices for line 1
prices2 = np.array([98, 99, 100, 103, 104, 106, 107, 109, 110, 115])  # Example stock prices for line 2
plt.plot(days, prices1, color='red')
plt.plot(days, prices2, color='green')

# Add labels
plt.xlabel('Days')
plt.ylabel('Prices, $')
plt.title('Stock Prices Over Time')
plt.show()
```

### Exercise 4 - Scatterplots

```python
# Plot price as a function of time
plt.scatter(days, prices, color='green', s=50)

# Show plot
plt.show()
```

### Exercise 4 - Is data normally distributed?

```python
# Plot histogram 
plt.hist(prices, bins=10)

# Display plot
plt.show()
```

### Exercise 5 - Comparing two histograms

```python
# Plot histogram of stocks_A
plt.hist(prices1, bins=10, alpha=0.4)

# Plot histogram of stocks_B 
plt.hist(prices2, bins=10, alpha=0.4)

# Display plot
plt.show()
```

### Exercise 6 - Adding a legend

```python
# Plot stock_A and stock_B histograms
plt.hist(prices1, bins=10, alpha=0.4, label='Stock A')
plt.hist(prices2, bins=10, alpha=0.4, label='Stock B')

# Add the legend
plt.legend()

# Display the plot
plt.show()
```
