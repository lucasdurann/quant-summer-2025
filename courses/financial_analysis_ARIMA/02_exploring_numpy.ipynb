```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random samples from a standard normal distribution (mean=0, std=1)
samples = np.random.normal(loc=0, scale=1, size=1000)

# Time Series Plot
plt.figure(figsize=(12, 4))
plt.plot(samples)
plt.title('Time Series of Standard Normal Samples')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Histogram Plot
plt.figure(figsize=(8, 4))
plt.hist(samples, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Standard Normal Samples')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Create a slight upward linear trend
trend = 0.002 * np.arange(len(samples))

# Add the trend to the original samples
samples_with_trend = samples + trend

# Time Series Plot with Trend Line
plt.figure(figsize=(12, 4))
plt.plot(samples_with_trend, label='Noisy Signal with Trend')
plt.plot(trend, color='red', linewidth=2, label='Trend Line')
plt.title('Time Series of Standard Normal Samples with Trend Line')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define parameters
mean = [0, 0]
cov = [[1, -0.5], [-0.5, 2]]  # Covariance matrix

# Generate 1000 samples
samples = np.random.multivariate_normal(mean, cov, size=1000)

# Scatter Plot
plt.figure(figsize=(6, 6))
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.6, edgecolor='k', s=30)
plt.title('Scatter Plot of Multivariate Normal Samples')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')  # Keeps aspect ratio to reflect true covariance
plt.show()
```

```python
# Manually compute sample mean vector
sample_mean = np.mean(samples, axis=0)

# Manually compute sample covariance matrix
centered_samples = samples - sample_mean
sample_cov = np.dot(centered_samples.T, centered_samples) / (samples.shape[0] - 1)

print("Sample Mean:\n", sample_mean)
print("\nSample Covariance Matrix:\n", sample_cov)
```
